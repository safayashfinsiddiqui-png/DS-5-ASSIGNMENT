```python
import pandas as pd
```

```python
df = pd.read_csv("sales.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
num_cols = df.select_dtypes(include='number').columns.tolist()
print("numeric columns:", num_cols)
```

```python
for col in num_cols:
    plt.figure()
    df.boxplot(column=col)
    plt.title(f"boxplot of {col}")
    plt.ylabel(col)
    plt.show()
```

```python
# categorical columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("categorical columns:", cat_cols)

      
```

```python

import matplotlib.pyplot as plt
TOP_N = 20
for col in cat_cols:
    counts = df[col].value_counts()

    print(f"\n{col} distribution:")
    print(counts.head(TOP_N))

    plt.figure()
    counts.head(TOP_N).plot(kind='bar')
    plt.title(f"Frequency of {col} (top {TOP_N})")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("numeric columns:", numeric_cols)
```

```python
print(df[numeric_cols].isna().sum())
print((df[numeric_cols].isna().sum() /
len(df)) * 100)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='median')
df_num = pd.DataFrame(imp.fit_transform(df[numeric_cols]),  columns=numeric_cols)
```

```python
plt.figure(figsize=(12,6))
df_num.hist(bins=30, figsize=(12,8))
plt.suptitle("Before standardization histogram")
plt.show()
```

```python
# Boxplot
for c in numeric_cols:
    plt.figure()
    df_num.boxplot(column=c)
    plt.title(f"Before standardiaztion: {c}")
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_std_vals = scaler.fit_transform(df_num)
df_std = pd.DataFrame(df_std_vals, columns=numeric_cols)
print(df_std.mean().round(6))   # should be approx 0
print(df_std.std().round(6))    # should be approx 1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure(figsize=(12,6))
df_std.hist(bins=30, figsize=(12,8))
plt.suptitle("After standardization - histograms")
plt.show()
```

```python
for c in numeric_cols:
    plt.figure()
    df_std.boxplot(column=c)
    plt.title(f"After standardization: {c}")
    plt.show()
```

```python
df_scaled = df.copy()
df_scaled[numeric_cols] = df_std
df_scaled.to_csv("data_standardization.csv", index=False)
print("saved standardized data as data_standardized.csv")
```
