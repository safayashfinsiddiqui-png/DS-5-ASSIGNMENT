1. Explain Bagging and Boosting methods. How is it different from each other.
1. Bagging and Boosting

Both Bagging and Boosting are ensemble learning methods — they combine multiple weak models (usually decision trees) to form a stronger overall model.
However, they differ in how they build and combine these models.
A. Bagging (Bootstrap Aggregating)

Concept:

Bagging builds multiple independent models in parallel.

Each model is trained on a random sample (with replacement) of the training data — called a bootstrap sample.

The final prediction is obtained by aggregating the outputs (e.g., taking the average for regression or majority vote for classification).

Key idea:
Reduce variance and avoid overfitting by averaging multiple models.

Example:

Random Forest is the most popular Bagging algorithm.
It trains multiple decision trees on different bootstrap samples and random subsets of features.
Steps:

Draw multiple bootstrap samples from the training data.

Train a base learner (e.g., decision tree) on each sample.

Aggregate their predictions (average or vote).

Advantages:

Reduces variance.

Parallelizable (since each model is built independently).

Works well with high-variance models like decision trees.

Disadvantages:

Doesn’t reduce bias much (since each model is trained independently).

Large number of models can be computationally expensive.
B. Boosting

Concept:

Boosting builds models sequentially, where each new model tries to correct the errors of the previous models.

It focuses more on bias reduction.

Key idea:
Train models sequentially so that each new model learns from the mistakes made by the previous ones.

Examples:

AdaBoost (Adaptive Boosting)

Gradient Boosting

XGBoost, LightGBM, CatBoost

Steps (AdaBoost example):

Start with equal weights for all data points.

Train a weak learner (e.g., shallow decision tree).

Increase the weights of misclassified samples so the next model focuses more on those.

Combine models with a weighted average to form the final strong model.
Advantages:

Reduces both bias and variance.

Can achieve very high accuracy.

Disadvantages:

Sensitive to noisy data and outliers.

Harder to parallelize (sequential process).

Can overfit if not tuned properly.

2. Explain how to handle imbalance in the data.
   2. Handling Imbalanced Data

Imbalanced data occurs when one class has far fewer samples than another (e.g., fraud detection, disease diagnosis).

A. Data-Level Methods

Resampling Techniques

Oversampling the minority class:
Duplicate or synthetically generate more minority samples.
Example: SMOTE (Synthetic Minority Oversampling Technique)

Undersampling the majority class:
Randomly remove samples from the majority class to balance the data.

Combination of both:
Apply both oversampling and undersampling for better results.

Data Augmentation:
For text or image data, create new samples by adding small variations (e.g., rotation, noise, translation).
B. Algorithm-Level Methods

Change the class weights:
Penalize misclassification of the minority class more heavily.
Many models (e.g., Logistic Regression, Random Forest, XGBoost) have a class_weight parameter.

Use specialized algorithms:

Balanced Random Forest

EasyEnsemble / BalancedBaggingClassifier

Cost-Sensitive Learning
C. Evaluation-Level Methods

Using regular accuracy can be misleading for imbalanced datasets.
Use metrics that focus on the minority class:

Metric	Description
Precision	Of predicted positives, how many are truly positive
Recall (Sensitivity)	Of actual positives, how many are correctly predicted
F1-score	Harmonic mean of Precision and Recall
ROC-AUC / PR-AUC	Evaluate model performance across different thresholds
Confusion Matrix	Shows True Positives, False Positives, etc.
